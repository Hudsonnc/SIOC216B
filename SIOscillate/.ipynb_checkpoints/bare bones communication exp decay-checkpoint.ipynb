```python
import numpy as np
from scipy.integrate import ode
import matplotlib as matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
class individual(object):
    def __init__(self, lambda_v=.1, lambda_a=.1,
                 v0=0, a0=0, 
                 varv=1, vara=1,
                 thresh=1
                ):
        self.lambda_v=lambda_v
        self.lambda_a=lambda_a
        self.y0=[v0, a0]
        self.varv=varv
        self.vara=vara
        self.thresh=thresh
        
    def f(self, t, y, force = None):
        v,a = y
        forcev = 0 if force is None else force['v']
        forcea = 0 if force is None else force['a']
        
        dvdt = -self.lambda_v*v + forcev
        dadt = -self.lambda_a*a + forcea
        return(np.array([dvdt, dadt]))
    
    def G(self, t, y):
        return(np.array([self.varv,self.vara]))
    
class population(object):
    def __init__(self, N=10):
        self.N = N
        self.individuals = [individual() for i in range(N)]
        v0 = [ind.y0[0] for ind in self.individuals]
        a0 = [ind.y0[1] for ind in self.individuals]
        self.y0 = np.hstack([v0, a0])
        
    def f(self, t, y, force = None):
        v = y[:self.N]
        a = y[self.N:]
        yzip = list(zip(v,a))
        
        force = [None for i in range(self.N)] if force is None else force
        
        dvdt, dadt = zip(*[self.individuals[i].f(t, yzip[i], force[i])
                           for i in range(self.N)
                          ])
        return(np.hstack([dvdt, dadt]))
    
    def G(self, t, y):
        v = y[:self.N]
        a = y[self.N:]
        yzip = list(zip(v,a))
        
        varv, vara = zip(*[self.individuals[i].G(t, yzip[i])
                           for i in range(self.N)
                          ])
        return(np.hstack([varv, vara]))
```

```python
def dW(n, dt, rng=np.random.RandomState()):
    return(rng.normal(size=n, loc=0, scale=np.sqrt(dt)))

def euler_maruyama_step(y_t, dt, f_t, G_t, rng=np.random.RandomState()):
    n = len(y_t)
    y_t_plus_1 = y_t + f_t*dt + G_t*dW(n, dt, rng=rng)
    return(y_t_plus_1)

class sde(object):
    def __init__(self, f, G, rng = np.random.RandomState()):
        self.f=f
        self.f_params=()
        self.G=G
        self.G_params=()
        self.rng=rng
        
    def set_initial_value(self, y, t=0):
        if np.isscalar(y):
            y = [y]
        self.y=np.array(y)
        self.t=t
        return(self)
    
    def set_f_params(self, *args):
        self.f_params=args
        return(self)
    
    def set_G_params(self, *args):
        self.f_params=args
        return(self)
    
    def step(self, dt):
        y = euler_maruyama_step(self.y, dt, 
                                self.f(self.t, self.y, *self.f_params),
                                self.G(self.t, self.y, *self.G_params),
                                rng=self.rng
                               )
        self.y = y
        self.t = self.t+dt
    
```

```python
y0 = np.zeros(10)
f = lambda t, x: -x
G = lambda t, x: 1

t0 = 0
dt = .1
T = 10

solver = sde(f, G)
solver.set_initial_value(y0)


t = [t0]
y = [y0]

while t[-1] < T:
    solver.step(dt)
    t.append(solver.t)
    y.append(solver.y)
    solver.set_f_params
    
y = np.array(y)
fig = plt.plot(t, y)
```

```python
ind = individual(lambda_v=1, lambda_a=1,
                 v0=0, a0=0, 
                 varv=.2, vara=1
                )

y0 = ind.y0
f = ind.f
G = ind.G

t0 = 0
dt = .01
T = 10

solver = sde(f, G)
solver.set_initial_value(y0)


t = [t0]
y = [y0]

while t[-1] < T:
    solver.step(dt)
    t.append(solver.t)
    y.append(solver.y)
    solver.set_f_params({'v': 1,'a': 0})
    
y = np.array(y)
fig = plt.plot(t, y)
```

```python
N = 2
pop = population(N)

y0 = pop.y0
v0 = y0[:N]
a0 = y0[N:]

f = pop.f
G = pop.G

t0 = 0
dt = .01
T = 100

solver = sde(f, G)
solver.set_initial_value(y0)


t = [t0]
v = [v0]
a = [a0]

while t[-1] < T:
    solver.step(dt)
    t.append(solver.t)
    _v = solver.y[:N]
    _a = solver.y[N:]
    v.append(_v)
    a.append(_a)
    
    forces = []
    for i in range(N):
        forces.append(
            {
                'v': np.tanh(np.mean([_v[j] if np.abs(_a[j]) > pop.individuals[j].thresh \
                             and j != i \
                             else 0
                             for j in range(N)
                            ])),
                'a': 0
            }
        )
    solver.set_f_params(forces)
        
    
v = np.array(v)
a = np.array(a)
plt.title('Valence')
figv = plt.plot(t, v)
plt.figure()
plt.title('Arousal')
figa = plt.plot(t, a)
plt.figure()
plt.title('Valence of 1 vs 2')
figvsc = plt.scatter(v[:,0], v[:,1])
plt.figure()
plt.title('Arousal of 1 vs 2')
figasc = plt.scatter(a[:,0], a[:,1])
```
